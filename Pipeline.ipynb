```python
from DFFScraper import DFFScraper
from GeneticAlgorithm import GeneticAlgorithm
```

```python
"?slate={slate_type}&tc={team_count}"

params = {
    "sport" : "nba",
    "platform" : "draftkings",
    "start_date" : "2021-06-03",
    "slate_type" : "", 
    "team_count" : ""
}

url = 'https://www.dailyfantasyfuel.com/{}/projections/{}/{}/'.format(params['sport'], params['platform'], params['start_date'])
```

```python
### E Step
scraper = DFFScraper(url, params)
raw = scraper.scrape()
master_player_list = scraper.toPlayerList(raw)
```

```python
## T Step
## add salary cap
## transformation filters 


def filterPlayerPool(filter_key, player_list):
    return list(filter(lambda player: float(player[filter_key]) > 0, player_list))

player_list = filterPlayerPool('ppg_proj', master_player_list)
```

```python
sports = {
    "NBA" : {
        "Positions" : ["PG", "SG", "SF", "PF", "C", "G", "F", "Util"], 
        "Salary_Cap" : 1000
    }, 
}
positions = sports["NBA"]["Positions"]
```

```python
### Load and Process
ga = GeneticAlgorithm(50, 1000, positions, 'ppg_proj', player_list)
ga.run()
```
